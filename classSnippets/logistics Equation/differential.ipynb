```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

k = 0.2
y0 = 5

# Define the derivative function
def dx_dt(x, t):
    return -k * x

# Define the time points
t = np.linspace(0, 10, 100)

# Solve the differential equation
x = odeint(dx_dt, y0, t)

# Plot the solution
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('x')
plt.title('Solution of dx/dt = -kx')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def logistic_equation(r, x):
    return r * x * (1 - x)

def plot_graph(r_values):
    xy = 0.02
    n_values = range(1, 101)

    for r in r_values:
        x_values = [xy]
        for n in n_values[1:]:
            x = logistic_equation(r, x_values[-1])
            x_values.append(x)
        
        plt.plot(n_values, x_values, label=f'r = {r}')
    
    plt.xlabel('n')
    plt.ylabel('x')
    plt.legend()
    plt.show()

r_values = [0.01, 0.1, 0.5, 1, 2.57, 3, 3.1, 3.15, 3.2]
plot_graph(r_values)
```
